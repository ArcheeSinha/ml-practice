```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 100, 100_000)
y = np.sqrt(x)

plt.plot(x, y, label = "y = sqrt(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y = sqrt(x)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y, label = "y = sin(x)", color = "blue")

plt.axvline(0, color='black', linewidth=1.5, linestyle='--', label = "x = 0")
plt.axhline(0, color='black', linewidth=1.5, linestyle='--', label = "y = 0")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y = sin(x)")
plt.grid(True)
plt.legend()

plt.show()
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.cos(x)

plt.plot(x, y, label = "y = cos(x)", color = "purple")
plt.axvline(0, color='black', linewidth=1.5, linestyle='--', label = "x = 0")
plt.axhline(0, color='black', linewidth=1.5, linestyle='--', label = "y = 0")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y = cos(x)")
plt.grid(True)
plt.legend()
plt.show()
```
