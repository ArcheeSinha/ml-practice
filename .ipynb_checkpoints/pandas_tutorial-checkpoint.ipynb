```python
s=pd.Series([1,2,3,4,5])
print(s)

#index
print(s[0])
print(s[2])
```

```python
s=pd.Series([11,23,34,45,53], index = ["Adam","Don","Bob","Maya","Charlie"])
print(s)

#label
print(s["Bob"])
```
