```python
import matplotlib.pyplot as plt
```

```python
expenses = ["Salaries","Rent","Marketing","R&D","Miscellaneous"]
amounts = [500, 150, 120, 100, 50]


plt.style.use('default')
plt.pie(amounts, labels = expenses)
# plt.style.available
plt.show()
```

```python
expenses = ["Salaries","Rent","Marketing","R&D","Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

# colors = ["red","blue","green","orange","purple"]

# plt.style.use('dark_background')
plt.pie(amounts, labels = expenses, autopct = '%1.1f%%')
# plt.style.available
plt.show()
```

```python
expenses = ["Salaries","Rent","Marketing","R&D","Miscellaneous"]
amounts = [500, 150, 120, 100, 50]



plt.pie(amounts, labels = expenses, autopct = '%1.1f%%',
        wedgeprops = {
            'edgecolor':'black',
            "linewidth": 2,
            "linestyle": "--",
            "alpha":0.5,
            "fill": False
            }
        )

plt.show()
```

```python
expenses = ["Salaries","Rent","Marketing","R&D","Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

explode = [0, 0, 0, 0.2, 0]

plt.pie(amounts, labels = expenses, autopct = '%1.1f%%',
        explode=explode, startangle=90, shadow = True
        )

plt.show()
```
