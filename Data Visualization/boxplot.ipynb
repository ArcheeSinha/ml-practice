```python
import matplotlib.pyplot as plt
```

```python
data = [7,8,5,6,9,4,10,12,15,30]

plt.boxplot(data, vert= False, showmeans= True, whis = 2.0)
plt.grid(True)
```

```python
#Multiple Datasets
import numpy as np

group1= np.random.normal(50,10,100)
group2= np.random.normal(60,15,100)

plt.boxplot([group1, group2], tick_labels = ['Group 1', 'Group 2'])
plt.grid(True)
plt.show()
```
