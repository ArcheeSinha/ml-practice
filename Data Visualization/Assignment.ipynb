```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
scores = np.random.normal (loc= 80, scale = 10, size =200)

mean = np.mean(scores)
plt.hist(scores, edgecolor = "black")
plt.axvline(mean, color = "red", linestyle = "--")
plt.title("Distribution of Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```

```python
penguins = sns.load_dataset ("penguins")

sns.scatterplot(
    data = penguins,
    x = "flipper_length_mm",
    y= "body_mass_g",
    hue = "species"
)
```

```python

tips = sns.load_dataset("tips")

sns.boxplot(
    data= tips,
    x = "day",
    y = "total_bill",
    hue = "sex"

)

median_bill = tips.groupby("day")["total_bill"].median()
print(median_bill)


# Highest Median Bill

# Sunday (Sun) has the highest median total bill.

# Comments on Outliers
# Outliers appear as points beyond the whiskers of the box plot.
# Saturday and Sunday generally show more outliers because larger groups often dine on weekends, leading to unusually high bills.
# These outliers indicate customers who spent significantly more than the typical customer for that day.
```

```python
tips = sns.load_dataset("tips")

ax = sns.scatterplot(
    data= tips,
    x = "total_bill",
    y = "tip",
    hue = "smoker",
    style = "sex"
    )

ax.set_title("Total Bill vs Tip by Gender")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
plt.legend(title = "Smoker")
```

```python
tips = sns.load_dataset("tips")

pivot_table = tips.pivot_table(
    values = "tip",
    index = "day",
    columns = "time",
    aggfunc = "mean"
)
print(pivot_table)

sns.heatmap(
    pivot_table,
    annot = True,
    cmap = "YlGnBu",
    fmt = ".2f"
)
```
