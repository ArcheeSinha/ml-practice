```python
# Real Data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]

oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

#Plot
plt.plot(years, oscar_revenue)
plt.title("Oscar Winner Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")
```

```python
# Non Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire", "Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2787, 829, 185, 275]  # in $M

plt.plot(years, oscar_revenue,color = "#CBC3E3",marker = "o", linestyle="-." ,linewidth = 5 ,label = "Oscar Movies")
plt.plot(years, non_oscar_revenue,">-.b", label = "Non Oscar Movies")

plt.title("Oscar Movies Revenue vs Non Oscar Movies Revenue ")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")

plt.legend(loc = "upper left")
```
