```python
# Solution 1
import requests
from bs4 import BeautifulSoup

URL = "http://quotes.toscrape.com/page/1/"
page_count = 1

while True:
    URL = f"http://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)
    
    soup = BeautifulSoup(res.text, "lxml")
    quotes = soup.select("div.quote")
    if not quotes:
        break

    with open(f"scraped_data/quotes{page_count}.html", "w") as f:
        f.write(res.text)
        print("downloaded data from page", page_count)

    page_count += 1
```

```python
# Solution 2 EXTRACT USEFUL INFO
with open (f"scraped_data/quotes1.html", "r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, "lxml")
all_quotes = soup.select("div.quote")
```

```python
all_quotes = soup.select("div.quote")
life_quotes = []
for q in all_quotes:
    all_tags =[]

    for tag in q.select(".tags .tag"):
        all_tags.append(tag.get_text())

    if "life" in all_tags:
        text = q.select_one("span.text").get_text()
        author = q.select_one("small.author").get_text()
        life_quotes.append((text, author))

print(life_quotes)
```
